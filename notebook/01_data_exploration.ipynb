## Data Loading

```python
# X
cpi = utils.load_raw('CPIAUCSL.csv') # Consumer Price Index - for All Urban Consumers: All Items in U.S. City Average
unrate = utils.load_raw('UNRATE.csv') # Unemployment Rate
infl = utils.load_raw('FPCPITOTLZGUSA.csv') # Inflation
gdp = utils.load_raw('GDPC1.csv') # Gross Domestic Product
disin = utils.load_raw('DSPIC96.csv') # Disposable Income
inprod = utils.load_raw('INDPRO.csv') # Industrial Production
claim = utils.load_raw('CCNSA.csv') # Continued Claims (Insured Unemployment)
leadingid = utils.load_raw('USSLIND.csv') # Leading Index for the United States

# Y
usrec = utils.load_raw('USRECP.csv') # Recession Indicator
```

## Left Join Data

```python
x = pd.merge(cpi, unrate, on='DATE', how = 'left')
x = pd.merge(x, infl, on='DATE', how = 'left')
x = pd.merge(x, gdp, on='DATE', how = 'left')
x = pd.merge(x, disin, on='DATE', how = 'left')
x = pd.merge(x, inprod, on='DATE', how = 'left')
x = pd.merge(x, claim, on='DATE', how = 'left')
x = pd.merge(x, leadingid, on='DATE', how = 'left')
```
